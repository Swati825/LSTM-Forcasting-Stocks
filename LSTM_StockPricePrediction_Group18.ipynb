### Part 4. Create Train, Validation, and Test Datasets

```python
import numpy as np
import datetime
import pandas as pd

def rssk3550_str_to_datetime(rssk3550_s):
    rssk3550_split = rssk3550_s.split('-')
    rssk3550_year, rssk3550_month, rssk3550_day = int(rssk3550_split[0]), int(rssk3550_split[1]), int(rssk3550_split[2])
    return datetime.datetime(year=rssk3550_year, month=rssk3550_month, day=rssk3550_day)

def rssk3550_df_to_windowed_df(rssk3550_dataframe, rssk3550_first_date_str, rssk3550_last_date_str, rssk3550_n=3):
    rssk3550_first_date = rssk3550_str_to_datetime(rssk3550_first_date_str)
    rssk3550_last_date = rssk3550_str_to_datetime(rssk3550_last_date_str)

    rssk3550_dates = []
    rssk3550_X, rssk3550_Y = [], []

    rssk3550_current_date = rssk3550_first_date

    while rssk3550_current_date <= rssk3550_last_date:
        rssk3550_window = []
        rssk3550_temp_date = rssk3550_current_date

        # Find the values for the previous n valid trading days
        for i in range(rssk3550_n):
            rssk3550_temp_date -= datetime.timedelta(days=1)
            while rssk3550_temp_date >= rssk3550_dataframe.index[0]:
                rssk3550_temp_date_dt = pd.to_datetime(rssk3550_temp_date)
                if rssk3550_temp_date_dt in pd.to_datetime(rssk3550_dataframe.index):
                    rssk3550_window.insert(0, rssk3550_dataframe.loc[rssk3550_temp_date_dt])
                    break
                rssk3550_temp_date -= datetime.timedelta(days=1)
            else: # Break out if we are out of dates
                break

        # Check if we have enough data for the window
        if len(rssk3550_window) == rssk3550_n and pd.to_datetime(rssk3550_current_date) in pd.to_datetime(rssk3550_dataframe.index):
            rssk3550_X.append(rssk3550_window)
            rssk3550_Y.append(rssk3550_dataframe.loc[pd.to_datetime(rssk3550_current_date)])
            rssk3550_dates.append(rssk3550_current_date)

        rssk3550_current_date += datetime.timedelta(days=1)

    rssk3550_ret_df = pd.DataFrame({})
    rssk3550_ret_df['Target Date'] = rssk3550_dates

    rssk3550_X = np.array(rssk3550_X)
    for rssk3550_i in range(0, rssk3550_n):
        rssk3550_ret_df[f'Target-{rssk3550_n-rssk3550_i}'] = rssk3550_X[:, rssk3550_i]

    rssk3550_ret_df['Target'] = rssk3550_Y

    return rssk3550_ret_df
```

```python
import pandas as pd
import numpy as np

# Load the windowed data from CSV
rssk3550_nifty50_df = pd.read_csv('nifty50_windowed_df.csv')
rssk3550_mahindra_df = pd.read_csv('mahindra_windowed_df.csv')

# Convert 'Target Date' column to datetime format
rssk3550_nifty50_df["Target Date"] = pd.to_datetime(rssk3550_nifty50_df["Target Date"])
rssk3550_mahindra_df["Target Date"] = pd.to_datetime(rssk3550_mahindra_df["Target Date"])

# Function to extract date, X (features), and y (target) from windowed DataFrame
def rssk3550_windowed_df_to_date_X_y(rssk3550_windowed_dataframe):
    rssk3550_dates = rssk3550_windowed_dataframe["Target Date"].to_numpy()  # Keep dates as datetime

    rssk3550_X = rssk3550_windowed_dataframe.iloc[:, 1:-1].to_numpy(dtype=np.float32)  # Extract features
    rssk3550_X = rssk3550_X.reshape((len(rssk3550_dates), rssk3550_X.shape[1], 1))  # Reshape

    rssk3550_Y = rssk3550_windowed_dataframe.iloc[:, -1].to_numpy(dtype=np.float32)  # Extract target values

    return rssk3550_dates, rssk3550_X, rssk3550_Y

# Extracting data for NIFTY 50
rssk3550_dates_nifty, rssk3550_X_nifty, rssk3550_y_nifty = rssk3550_windowed_df_to_date_X_y(rssk3550_nifty50_df)
print(f"NIFTY 50 Data Shapes - Dates: {rssk3550_dates_nifty.shape}, X: {rssk3550_X_nifty.shape}, Y: {rssk3550_y_nifty.shape}")

# Extracting data for Mahindra & Mahindra
rssk3550_dates_mahindra, rssk3550_X_mahindra, rssk3550_y_mahindra = rssk3550_windowed_df_to_date_X_y(rssk3550_mahindra_df)
print(f"Mahindra Data Shapes - Dates: {rssk3550_dates_mahindra.shape}, X: {rssk3550_X_mahindra.shape}, Y: {rssk3550_y_mahindra.shape}")
```

**Scaling the Data with MinMaxScaler**

```python
from sklearn.preprocessing import MinMaxScaler

# Create a scaler object for Nifty
rssk3550_scaler_nifty = MinMaxScaler()

# Reshape the data for scaling (if necessary)
rssk3550_X_train_nifty_reshaped = rssk3550_X_train_nifty.reshape(-1, 1)
rssk3550_y_train_nifty_reshaped = rssk3550_y_train_nifty.reshape(-1, 1)

# Fit the scaler to the training data and transform it
rssk3550_X_train_nifty_scaled = rssk3550_scaler_nifty.fit_transform(rssk3550_X_train_nifty_reshaped).reshape(rssk3550_X_train_nifty.shape)
rssk3550_y_train_nifty_scaled = rssk3550_scaler_nifty.fit_transform(rssk3550_y_train_nifty_reshaped).reshape(rssk3550_y_train_nifty.shape)

# Scale validation and testing data using the fitted scaler
rssk3550_X_val_nifty_scaled = rssk3550_scaler_nifty.transform(rssk3550_X_val_nifty.reshape(-1, 1)).reshape(rssk3550_X_val_nifty.shape)
rssk3550_y_val_nifty_scaled = rssk3550_scaler_nifty.transform(rssk3550_y_val_nifty.reshape(-1, 1)).reshape(rssk3550_y_val_nifty.shape)

rssk3550_X_test_nifty_scaled = rssk3550_scaler_nifty.transform(rssk3550_X_test_nifty.reshape(-1, 1)).reshape(rssk3550_X_test_nifty.shape)
rssk3550_y_test_nifty_scaled = rssk3550_scaler_nifty.transform(rssk3550_y_test_nifty.reshape(-1, 1)).reshape(rssk3550_y_test_nifty.shape)
```

```python
# Model for Nifty50
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import layers
from tensorflow.keras.callbacks import ReduceLROnPlateau

rssk3550_model_nifty50 = Sequential([
    layers.Input((3, 1)),
    layers.LSTM(128, return_sequences=True),  # Increased units, return sequences for stacked LSTM
    layers.LSTM(64),  # Added another LSTM layer
    layers.Dense(32, activation='relu'),
    layers.Dropout(0.2),  # Added dropout for regularization
    layers.Dense(32, activation='relu'),
    layers.Dense(1)
])

# Learning rate scheduler
rssk3550_lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.0001)

rssk3550_model_nifty50.compile(loss='mse',
                    optimizer=Adam(learning_rate=0.001),
                    metrics=['mean_absolute_error'])

rssk3550_model_nifty50.fit(rssk3550_X_train_nifty_scaled, rssk3550_y_train_nifty_scaled,
                    validation_data=(rssk3550_X_val_nifty_scaled, rssk3550_y_val_nifty_scaled),
                    epochs=200,
                    callbacks=[rssk3550_lr_scheduler])
```

```python
# Model for Mahindra
rssk3550_model_mahindra = Sequential([
    layers.Input((3, 1)),
    layers.LSTM(64),
    layers.Dense(32, activation='relu'),
    layers.Dense(32, activation='relu'),
    layers.Dense(1)
])

rssk3550_model_mahindra.compile(loss='mse',
                    optimizer=Adam(learning_rate=0.001),
                    metrics=['mean_absolute_error'])

rssk3550_model_mahindra.fit(rssk3550_X_train_mahindra, rssk3550_y_train_mahindra,
                    validation_data=(rssk3550_X_val_mahindra, rssk3550_y_val_mahindra),
                    epochs=200)
```

```python
# For Nifty50
from copy import deepcopy
import numpy as np

rssk3550_recursive_predictions_nifty50 = []
rssk3550_recursive_dates_nifty50 = np.concatenate([rssk3550_dates_val_nifty, rssk3550_dates_test_nifty]) # Changed rssk3550_dates_val_nifty50 to rssk3550_dates_val_nifty

for target_date in rssk3550_recursive_dates_nifty50:
    last_window = deepcopy(rssk3550_X_train_nifty_scaled[-1])  # Get the last sequence of the scaled training data
    next_prediction = rssk3550_model_nifty50.predict(np.array([last_window])).flatten()[0]  # Predict next value
    rssk3550_recursive_predictions_nifty50.append(next_prediction)

    # Properly update the window by shifting left and appending new prediction
    last_window = np.roll(last_window, shift=-1, axis=0)
    last_window[-1, 0] = next_prediction  # Insert new prediction at last position

# For Mahindra
rssk3550_recursive_predictions_mahindra = []
rssk3550_recursive_dates_mahindra = np.concatenate([rssk3550_dates_val_mahindra, rssk3550_dates_test_mahindra])

for target_date in rssk3550_recursive_dates_mahindra:
    last_window = deepcopy(rssk3550_X_train_mahindra[-1])
    next_prediction = rssk3550_model_mahindra.predict(np.array([last_window])).flatten()[0]
    rssk3550_recursive_predictions_mahindra.append(next_prediction)

    last_window = np.roll(last_window, shift=-1, axis=0)
    last_window[-1, 0] = next_prediction
```

```python
# Model for Nifty50
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import layers
from tensorflow.keras.callbacks import ReduceLROnPlateau

rssk3550_model_nifty50 = Sequential([
    layers.Input((3, 1)),
    layers.LSTM(128, return_sequences=True),  # Increased units, return sequences for stacked LSTM
    layers.LSTM(64),  # Added another LSTM layer
    layers.Dense(32, activation='relu'),
    layers.Dropout(0.2),  # Added dropout for regularization
    layers.Dense(32, activation='relu'),
    layers.Dense(1)
])

# Learning rate scheduler
rssk3550_lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.0001)

rssk3550_model_nifty50.compile(loss='mse',
                    optimizer=Adam(learning_rate=0.001),
                    metrics=['mean_absolute_error'])

# Assign the fit method to rssk3550_history_nifty50
rssk3550_history_nifty50 = rssk3550_model_nifty50.fit(rssk3550_X_train_nifty_scaled, rssk3550_y_train_nifty_scaled,
                    validation_data=(rssk3550_X_val_nifty_scaled, rssk3550_y_val_nifty_scaled),
                    epochs=50,
                    callbacks=[rssk3550_lr_scheduler])
```

```python
# Model for Mahindra
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import layers
from tensorflow.keras.callbacks import ReduceLROnPlateau # Importing ReduceLROnPlateau

rssk3550_model_mahindra = Sequential([
    layers.Input((3, 1)),
    layers.LSTM(64),
    layers.Dense(32, activation='relu'),
    layers.Dense(32, activation='relu'),
    layers.Dense(1)
])

rssk3550_model_mahindra.compile(loss='mse',
                    optimizer=Adam(learning_rate=0.001),
                    metrics=['mean_absolute_error'])

# Assign the fit method to rssk3550_history_mahindra, similar to Nifty50
rssk3550_history_mahindra = rssk3550_model_mahindra.fit(rssk3550_X_train_mahindra, rssk3550_y_train_mahindra,
                    validation_data=(rssk3550_X_val_mahindra, rssk3550_y_val_mahindra),
                    epochs=50)
```
